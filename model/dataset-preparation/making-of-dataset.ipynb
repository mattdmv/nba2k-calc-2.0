```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

## Player rating NBA2k22

```python
nba2k22_ratings = pd.read_csv('nba2k22_ratings.csv', sep=';')
```

```python
nba2k22_ratings.sort_values('PLAYER', inplace=True)
```

```python
nba2k22_ratings['PLAYER'] = nba2k22_ratings['PLAYER'].str.rstrip()
```

```python
nba2k22_ratings.info()
```

## Player stats for season 2020/21

```python
stats2021 = pd.read_csv('nba_stats_2021.csv', sep=';')
```

```python
stats2021
```

```python
stats2021.shape
```

```python
stats2021.info()
```

## Merge dataframes and change datatypes

```python
nba_ratings_2022 = stats2021.merge(nba2k22_ratings, how='inner', on='PLAYER')
```

```python
nba_ratings_2022
```

```python
nba_ratings_2022.info()
```

```python
nba_ratings_2022['rankings'] = nba_ratings_2022['rankings'].astype(float)
```

```python
nba_ratings_2022.info()
```

```python
# read in 2014-2020 dataset
nba_ratings_2014_2020 = pd.read_csv('nba_rankings_2014-2020.csv')
```

```python
nba_ratings_2014_2020.head(20)
```

```python
nba_ratings_2014_2020['rankings'] = nba_ratings_2014_2020['rankings'].astype(int)
```

```python
nba_ratings_2014_2020.info()
```

```python
# drop excess columns
cols_to_drop = ['Unnamed: 0', 'SEASON']
```

```python
nba_ratings_2014_2020.drop(cols_to_drop, axis=1, inplace=True)
```

```python
nba_ratings_2014_2020
```

## Concat dataframes into final

```python
# check columns
nba_ratings_2014_2020.columns == nba_ratings_2022.columns
```

```python
final = pd.concat([nba_ratings_2014_2020, nba_ratings_2022], axis=0)
```

```python
final.shape
```

```python
final
```

```python
#reset index
final.reset_index(drop=True, inplace=True)
```

```python
final
```

```python
final.info()
```

```python
final.drop(['PLAYER', 'TEAM'], axis=1, inplace=True)
```

```python
final.head(5)
```

```python
# save final dataframe to csv file
final.to_csv('../dataset.csv')
```
